```python
import math as m

phi0=lambda x: m.sin(2*m.pi*x)+m.sin(20*m.pi*x)+m.sin(64*m.pi*x) #the initial state of the system
n=257
k=100

fig, ax= plt.subplots(ncols=3, nrows=2, sharey=True, figsize=(20,12))
ax[0][0].set_ylim(-3,3)

PJ(phi0,lambda x:0,n,0,1,k, ax=ax[0][0]);
GS(phi0,lambda x:0,n,0,1,k, ax=ax[0][1]);
SOR(phi0,lambda x:0,n,0,1,k,ax=ax[0][2]);
DG(phi0,n,0,1,k, ax=ax[1][0]);
MGv(phi0,n,0,1,k, ax=ax[1][1]);
```
